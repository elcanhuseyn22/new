### Videoların bəyənmə nisbətinin ən çoxundan ən aza doğru sıralanması

```python
def likesdislakesratio(likes,dislakes):
    likesList = []
    for key,value in list(likes.iteritems()):
        likesList.append(value)
    dislikesList = []
    for key,value in list(dislakes.iteritems()):
        dislikesList.append(value)
    
    likedislike = list(zip(likesList,dislikesList))
    ratio = []
    for like,dislike in likedislike:
        if like+dislike == 0:
            ratio.append(0)
        else:
            ratio.append(like / (like+dislike))

    return ratio
```

```python
likesdislakesratio(youtube["likes"],youtube["dislikes"])
```

```python
youtube["likes_dislikes"] = likesdislakesratio(youtube["likes"],youtube["dislikes"])
```

```python
youtube.sort_values("likes_dislikes",ascending = False, inplace = True) 
```
